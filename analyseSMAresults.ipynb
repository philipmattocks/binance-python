```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

```

```python
def plot_sma_results(csv_file, first_n_rows = 0):

        
    #plt.scatter(df['x'], df['sma5'],color='red')
    df = pd.read_csv(csv_file)
    if first_n_rows > 0:
        df = df[:first_n_rows]
    fig = plt.figure()
    ax = fig.add_subplot(1, 1, 1)
    ax.xaxis.set_major_locator(plt.MaxNLocator(20))
    plt.rcParams["figure.figsize"] = (20,3)
    plt.plot( 'date', 'sma_50', data=df, color='red', linewidth=1)
    plt.plot( 'date', 'sma_200', data=df, color='blue', linewidth=1)
    plt.plot( 'date', 'price', data=df, color='green', linewidth=1)
    plt.scatter(df['date'], df['long'], color='red',s=50)
    plt.scatter(df['date'], df['short'], color='blue',s=50)
    plt.xticks(rotation=45)
    plt.show()
```
